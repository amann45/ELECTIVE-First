##### Types of argument in function: positional argument, keyword argument, variable length argument(args*,kwargs**)

```python
def sum(a=6,b=10):
    print("the value of a is",a)
    print("the value of b is",b)

sum(20,50)
```

```python
def myfun(**kwargs):
        for k,v in kwargs.items():
            print(f"the value if key is {k} and values is{v}")

myfun(a=10,b=40,c=50)
```

#### lambda functions

```python

a=lambda a:a+2
print(a(4))
```

```python
l=(0,5,6,7,8,9,10)
l1=tuple(filter(lambda x:x%2==0, l))
print(l1)
```

```python
from functools import reduce
l=(0,5,6,7,8,9,10)
l1=reduce(lambda x,y:x+y, l)
print(l1)
```

### function Decorators:

```python
def decor(a):
    def inner(name):
          if name=="ajay":
               print("known person")
          else:
               print("i dont know your name")
    return inner

@decor
def fun(name):
    if name== "ajay":
        print("ajay")
    else:
            print("guest")
fun("ajay")
```
