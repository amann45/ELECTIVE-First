# if and loops conditions

```python
# if condition
a=int (input("enter any a number:"))
b=int (input("enter any b number:"))
if a>b: 
    print("this is true condition")
else:     #else condition
        print("this is false condition")
```

```python

# if elif else
name=input("enter any name:")
if name=="aman":
 print("hello aman")
elif name=="aarogya":
   print("hello aarogya")
else:
    print("hello guest")
```

```python
# for loop
# for x in sequence: list, tuples, range, set, string
list=[10,20,30,50]
for x in list:
    print(x)
```

```python
#while loop
x=0
while (x<10):
    print(f"value of x is {x}")
    x=x+1
```

```python
for i in range(5):
    for j in range(5):
        print(f"The value of i is {i} and the value of j is {j}")
```
